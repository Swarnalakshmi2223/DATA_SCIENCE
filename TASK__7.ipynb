**1).Make a three-dimensional plot with randomly generate 50 data points for x, y, and z. Set

**4 bold text**).i. Using Pandas, Create a DataFrame with a list of dictionaries, row indices, and column
indices.


ii. Use index label to delete or drop rows from a Pandas DataFrame.**

```python
import pandas as pd

# Create a list of dictionaries
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'San Francisco'},
    {'Name': 'Charlie', 'Age': 22, 'City': 'Los Angeles'}
]

# Define row indices
row_indices = ['A', 'B', 'C']

# Define column indices
column_indices = ['Name', 'Age', 'City']

# Create a DataFrame
df = pd.DataFrame(data, index=row_indices, columns=column_indices)
print(df)
```

```python
# Drop a specific row by index label
df = df.drop('B')

# Display the updated DataFrame
print(df)
```

Apply and explore various plotting functions on Pima Indians Diabetes data set for
performing the following:

a) Normal values

b) Density and contour plots

c) Three-dimensional plotting

**the Pima Indians Diabetes dataset**

```python
data=pd.read_csv('/content/gdrive/MyDrive/diabetes.csv')
print(data)
```

**importing libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

**a) Normal values - Histogram for a single variable**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Glucose'], kde=True, color='skyblue')
plt.title('Distribution of Glucose Levels')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.show()
```

**b) Density and contour plots - Pairplot for multiple variables**

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.pairplot(data, hue='Outcome', diag_kind='kde', markers=['o', 's'], palette='husl')
plt.suptitle('Pairplot of Pima Indians Diabetes Dataset', y=1.02)
plt.show()
```

**c) Three-dimensional plotting - Scatter plot in 3D**

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Glucose'], data['BMI'], data['Age'], c=data['Outcome'], cmap='viridis', marker='o')
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('3D Scatter Plot of Pima Indians Diabetes Dataset')
plt.show()
```

Use the diabetes data set from UCI data set for performing the following:

Apply Univariate analysis:

Frequency Mean,  Median,  Mode,  Variance  Standard Deviation
 Skewness and Kurtosis

```python
pip install pandas numpy scipy
```

**importing libraries**

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
```

**UCI DATASET**

```python
data = pd.read_csv('/content/gdrive/MyDrive/diabetes.csv')
print(data)
```

**Frequency analysis**

```python
frequency_count = data['Outcome'].value_counts()
print("Frequency of Outcome (0: No Diabetes, 1: Diabetes):")
print(frequency_count)
```

**MEAN,MEDIAN,MODE**

```python
mean_values=data.mean()
median_values=data.median()
mode_values=data.mode().iloc[0]
print("\nmean values:")
print(mean_values)
print("\nmedian values:")
print(median_values)
print("\nmode values:")
print(mode_values)
```

**Variance and Standard Deviation**

```python
variance_values = data.var()
std_deviation_values = data.std()

print("\nVariance Values:")
print(variance_values)

print("\nStandard Deviation Values:")
print(std_deviation_values)
```

**Skewness and Kurtosis **

```python
skewness_values = data.skew()
kurtosis_values = data.kurtosis()

print("\nSkewness Values:")
print(skewness_values)

print("\nKurtosis Values:")
print(kurtosis_values)
```
